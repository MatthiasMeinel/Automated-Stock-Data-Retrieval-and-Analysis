```python
#In this project we will automatize the process of gathering the stock data.
# To do this, we will use selenium and some of its modules to perform some webscrapping


import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
```

```python
#Let's now get the driver 

driver = webdriver.Chrome(service = Service(ChromeDriverManager().install()))

url = 'https://fundamentus.com.br/resultado.php'

driver.get(url)
```

```python
#Let's use our driver to get the html file

table_location = '/html/body/div[1]/div[2]/table'

element = driver.find_element('xpath', table_location)

```

```python
html_table = element.get_attribute("outerHTML")
```

```python
#Turning the file into a pandas table
#Notice that we are using the brazilian decimal system. For thousands we use '.' and for decimals ','

df = pd.read_html(str(html_table), thousands='.', decimal= ',')[0]
```

```python
df
```

```python
#To make the manipulation easier, we are going to turn the tickers into the index (in this table, Papel = ticker)

df = df.set_index('Papel')
```

```python
df.head()
```

```python
#Now we are going to choose the features we are going to use in our analysis

df = df[['Cotação', 'EV/EBIT','ROIC', 'Liq.2meses' ]]
```

```python
df.head()
```

```python
#Notice that we have a problem. We need to delete the % symbols and also structure the thousands and decimals symbols to the American pattern

df['ROIC']=df['ROIC'].str.replace('%', '')
df['ROIC']=df['ROIC'].str.replace('.', '')
df['ROIC']=df['ROIC'].str.replace(',', '.')
df['ROIC']=df['ROIC'].astype(float)

df
```

```python
#Now, with our data ready, let's add our filters to select the stocks 

df = df[df['Liq.2meses']>1000000]
df = df[df['EV/EBIT']>0]
df = df[df['ROIC']>0]
```

```python
df
```

```python
df['ranking_EV/EBIT'] = df['EV/EBIT'].rank(ascending=True)
df['ranking_ROIC'] = df['ROIC'].rank(ascending=False)
df['ranking_total'] = df['ranking_EV/EBIT'] + df['ranking_ROIC']

df.sort_values("ranking_total").head(20)
```
